```python
cf.set_indicators(trading_data, 200)
trading_data
```

```python
def total_trade_signal(row):
    if (row.MACD_trade_signal == 1):
        return 1
    elif (row.MACD_trade_signal == -1):
        return -1
    else:
        return 0

cf.set_trade_signals(trading_data, total_trade_signal)
trading_data
```

```python
backtest, stats, heatmap = cf.backtest_trading(trading_data, 25000)

print(stats)
backtest.plot()
cf.show_heatmap(heatmap)
```
