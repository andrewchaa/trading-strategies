```python
import common_functions as cf
import common_trade_signals as cts

trading_data = cf.read_data('BTC-USD', '1y', '1h')

cf.set_indicators(trading_data, ema_length=200)
cf.set_trade_signals(trading_data, cts.total_macd_trade_signal)

latest_data = trading_data.tail(2500)
latest_data
backtest, stats, heatmap = cf.backtest_trading(latest_data, 20000)

print(stats)
cf.show_heatmap(heatmap)
backtest.plot()
```
